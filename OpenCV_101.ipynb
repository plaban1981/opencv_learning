#### Import packages

```python
import matplotlib.pyplot as plt
import argparse
import cv2
```

#### Function to display images

* The **cv2.imread** function accepts a single parameter <**the path to where your image lives on your disk**>

* The OpenCV cv2.imread function then returns either of two values:

  * A NumPy array representing the image with the shape (num_rows, num_cols, num_channels), which we’ll discuss later in this tutorial
  * A NoneType object, implying that the image could not be loaded

* Typically, the cv2.imread function will return **None** if the path to the input image is invalid, 

```python
def plt_imshow(image,title):
  image = cv2.imread(image)
  image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
  plt.imshow(image)
  plt.title(title)
  plt.grid(False)
  plt.show()
```

```python
plt_imshow("/content/opencv-load-image/jurassic_park.png",'Jurrasic Park')
```
